```python
class Node:
    def __init__(self,value): # initialisation of a new node
        self.data = value
        self.next = None

    
```

```python
class LinkedList :
    def __init__(self):
        # empty linked list
        self.head = None
        self.n = 0 # number of nodes in LL

    def __len__(self):  # return number of nodes
        return self.n 
    
    def insert_head(self,val): # insert at beginning
        # create a new node from Node class
        new = Node(val)
        #create connection
        new.next = self.head
        self.head = new

        # increment n
        self.n += 1

    def __str__(self): # to traverse and print the elements of linked List
        curr = self.head # start at head
        result = ''
        while curr != None :
            result += str(curr.data) + '->'
            curr = curr.next        
        return  result[:-2]
    
    def insert_tail(self,val): # insert at end
        new = Node(val)
        if self.head == None:
            self.head = new
            self.n += 1
            return
        curr = self.head
        while curr.next != None :
            curr = curr.next # at last node stops
        curr.next = new
        self.n += 1

    def insert_after(self,after,val): # insert after a given value
        new = Node(val)
        curr = self.head
        while curr != None:
            if curr.data == after :
                break
            curr = curr.next
        # case 1 -> item found
        if curr != None:
            new.next = curr.next
            curr.next = new
            self.n += 1
        else: # case 2 -> item not found
            return "Item Not found"
        
    def clear(self): # clearing the whole linked list
        self.head = None
        self.n = 0

    def delete_head(self): # deleting from beginning
        if self.n > 0 :
            self.head = self.head.next
            self.n -= 1
        else :
            return "Empty linked list"
        
    def pop_list(self): # delete from tail
        curr = self.head
        if self.head == None :
            return "Empty linked list"
        
        if curr.next == None :
            self.delete_head()
            return
        
                    
        while curr.next.next != None :
            curr = curr.next
        curr.next = None
        self.n -= 1

    def remove_val(self,val):
        if self.head == None :
            return "Empty list"
        if self.head.data == val :
            self.delete_head()
            return
        curr = self.head
        while curr.next != None :
            if curr.next.data == val :
                break
            else :
                curr = curr.next
        if curr.next == None :
            return "Item Not Found"
        else:
            curr.next = curr.next.next

    def search(self,val):
        if self.head == None :
            return "Empty list"
        curr = self.head
        pos = 0

        while curr != None :
            if curr.data == val :
                return pos
            curr = curr.next
            pos += 1

        return "Not Found"
    
    def __getitem__(self,index):
        if self.head == None :
            return "Empty list"
        curr = self.head
        pos = 0 
        while curr != None:
            if pos == index :
                return curr.data
            curr = curr.next
            pos += 1
        return "Not found"

    #write a program to find the maximum value in a linked list and replace it with a given value  
    def replace_max(self,value):
        if self.head is None:
            return
        
        curr = self.head
        max = curr
        while curr != None :
            if curr.data > max.data :
                max = curr
            curr = curr.next 
        max.data = value
        
    # to find the sum of all items in odd position
    def sum_odd(self):
        if self.head is None:
            return
        
        curr = self.head
        pos = 0
        sum = 0
        while curr != None :
            if pos % 2 != 0:
                sum += curr.data
            pos += 1
            curr = curr.next 
        return sum
    
    # to reverse a linked list containing integer data

    def reverse(self):
        prev = None
        curr = self.head

        while curr != None:
            next = curr.next # store the next node in variable
            curr.next = prev # current node points towards previous node
            prev = curr
            curr = next
        self.head = prev
        
    '''Q. Given a linked list of characters write a program to return 
        a new string by appending all the characters as per the given rules :-
        1)replace * or / by a single space
        2)incase two consecutive * or / replace bt single space and convert
        next character to uppercase , assuming there are not more than two 
        consequtive appearance of * or / and list always ends with an alphabet
    '''

    def change_char(self):
        temp = self.head
        while temp != None:
            if temp.data == '*' or temp.data == '/' :
                temp.data = ' '
                if temp.next.data == '*' or temp.next.data == '/' :
                    temp.next.next.data = temp.next.next.data.upper()
                    temp.next = temp.next.next # character shifting

            temp = temp.next
     
    



    



        



```
