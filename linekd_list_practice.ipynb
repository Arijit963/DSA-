```python
class Node:
    def __init__(self,value): # initialisation of a new node
        self.data = value
        self.next = None

    
```

```python
class LinkedList :
    def __init__(self):
        # empty linked list
        self.head = None
        self.n = 0 # number of nodes in LL

    def __len__(self):  # return number of nodes
        return self.n 
    
    def insert_head(self,val): # insert at beginning
        # create a new node from Node class
        new = Node(val)
        #create connection
        new.next = self.head
        self.head = new

        # increment n
        self.n += 1

    def __str__(self): # to traverse and print the elements of linked List
        curr = self.head # start at head
        result = ''
        while curr != None :
            result += str(curr.data) + '->'
            curr = curr.next        
        return result[:-2]
    
    def insert_tail(self,val): # insert at end
        new = Node(val)
        if self.head == None:
            self.head = new
            self.n += 1
            return
        curr = self.head
        while curr.next != None :
            curr = curr.next # at last node stops
        curr.next = new
        self.n += 1

    def insert_after(self,after,val): # insert after a given value
        new = Node(val)
        curr = self.head
        while curr != None:
            if curr.data == after :
                break
            curr = curr.next
        # case 1 -> item found
        if curr != None:
            new.next = curr.next
            curr.next = new
            self.n += 1
        else: # case 2 -> item not found
            return "Item Not found"
        
    def clear(self): # clearing the whole linked list
        self.head = None
        self.n = 0

    def delete_head(self): # deleting from beginning
        if self.n > 0 :
            self.head = self.head.next
            self.n -= 1
        else :
            return "Empty linked list"
        
    def pop_list(self): # delete from tail
        curr = self.head
        if self.head == None :
            return "Empty linked list"
        
        if curr.next == None :
            self.delete_head()
            return
        
                    
        while curr.next.next != None :
            curr = curr.next
        curr.next = None
        self.n -= 1

    def remove_val(self,val):
        if self.head == None :
            return "Empty list"
        if self.head.data == val :
            self.delete_head()
            return
        curr = self.head
        while curr.next != None :
            if curr.next.data == val :
                break
            else :
                curr = curr.next
        if curr.next == None :
            return "Item Not Found"
        else:
            curr.next = curr.next.next

    def search(self,val):
        if self.head == None :
            return "Empty list"
        curr = self.head
        pos = 0

        while curr != None :
            if curr.data == val :
                return pos
            curr = curr.next
            pos += 1

        return "Not Found"
    
    def __getitem__(self,index):
        if self.head == None :
            return "Empty list"
        curr = self.head
        pos = 0 
        while curr != None:
            if pos == index :
                return curr.data
            curr = curr.next
            pos += 1
        return "Not found"
        

        



```
