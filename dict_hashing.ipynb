Linear Probing

```python
class Dictionary :
    def __init__(self,size):
        self.size = size
        self.slots = [None] * self.size # array for keys
        self.data = [None] * self.size # array for values

    def put(self, key, value):
        hash_val = self.hash_function(key)

        if self.slots[hash_val] == None :
            self.slots[hash_val] = key
            self.data[hash_val] = value
        else :

            if self.slots[hash_val] == key :
                self.data[hash_val] = value
            else : 
                new_hash = self.rehash(hash_val)

                while self.slots[new_hash] != None and self.slots[new_hash] != key :
                    new_hash = self.rehash(new_hash)

                if self.slots[new_hash] == None :
                    self.slots[new_hash] = key
                    self.data[new_hash] = value
                else :
                    self.data[new_hash] = value

    def __setitem__(self,key,value): # magic method for dict notation
        self.put(key,value)

    def get(self,key):
        start_pos = self.hash_function(key)
        curr = start_pos
        while self.slots[curr] != None :
            if self.slots[curr] == key :
                return self.data[curr]
            
            curr = self.rehash(curr)

            if curr == start_pos :
                return "Item not found"
            
        return "Not found"

    def __getitem__(self,key): # magic method for retrieving item
        return self.get(key)
    
    def __str__(self):
        for i in range(len(self.slots)):
            if self.slots[i] != None:
                print(self.slots[i] , " : " , self.data[i])

        return ""

    def rehash(self,old_hash):
        return (old_hash + 1) % self.size 


    def hash_function(self,key):
        return abs(hash(key)) % self.size

    
```

```python
d1 = Dictionary(3)
```
