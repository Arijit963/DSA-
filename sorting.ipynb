```python
def is_sorted(arr):
    sorted = True
    for i in range(len(arr)-1):
        if arr[i] > arr[i+1]:
            sorted = False
    return sorted
```

```python
a = [1,3,4,6,8,8]
is_sorted(a)
```

```python
# Monkey sort 
# time complexity -> O(infinity)
import random

def monkey_sort(arr):
    while not is_sorted(arr) :
        random.shuffle(arr)
        print(arr)
    print(arr)
```
