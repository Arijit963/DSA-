```python
import ctypes
```

```python
class MyList :
    def __init__(self):
        self.size = 1 # total size
        self.n = 0 # number of current elements
        self.A = self.__make_array(self.size) # list creation 
    
    def __len__(self): # magic method for len()
        return self.n
    
    def __resize(self,new_size): # resizing for dynamic array
        B = self.__make_array(new_size)
        self.size = new_size
        # we have to copy the content of the previous array
        for i in range(self.n):
            B[i] = self.A[i]
        # reassigning B to A
        self.A = B 

    def append(self,item): # to append to a list
        if self.n == self.size : # to resize if array is full
            self.__resize(self.size*2) 
        self.A[self.n] = item
        self.n += 1

    def __str__(self): # magic method used to print()
        result = ''
        for i in range(self.n):
            result = result + str(self.A[i]) + ','
        return '['+ result[:-1] +']'

    def __getitem__(self,index): # for indexing of list
        if 0 <= index < self.n :
            return self.A[index]
        else:
            return 'IndexError'
        
    def pop(self): # to pop the last element
        if self.n == 0:
            return 'Empty list'
        print(self.A[self.n -1])
        self.n = self.n - 1

    def clear(self): # to clear the whole list
        self.A = 1
        self.n = 0

    def find(self,item): # to find the index number of the item
        for i in range(self.n):
            if self.A[i] == item:
                return i
        return "Not in list"

    def insert(self,pos,item): # to insert an element in the position
        if self.n == self.size :
            self.__resize(self.size + 1)
        for i in range(self.n,pos,-1):
            self.A[i] = self.A[i-1]
        self.A[pos] = item
        self.n += 1

    def __delitem__(self,pos): # to delete an item from a position
        if 0 < pos < self.n :
           
            for i in range(pos,self.n - 1 ):
                self.A[i] = self.A[i+1]
            self.n -= 1

    def remove(self,item):
        pos = self.find(item)
        if type(pos) == int :
            self.__delitem__(pos)
        else:
            return "Item not in list"

    def min(self): # finding the minimum in the list
        if self.n == 0:
            return "Empty list"
        min_val = self.A[0]
        
        for i in range(1,self.n):
            if self.A[i] < min_val :
                min_val = self.A[i]
        return min_val
    
    def max(self): # finding the maximum in the list
        if self.n == 0:
            return "Empty list"
        max_val = self.A[0]
        
        for i in range(1,self.n):
            if self.A[i] > max_val :
                max_val = self.A[i]
        return max_val
    
    def sort(self):
        for i in range(self.n - 1):
            for j in range(self.n - i - 1):
                if self.A[j] > self.A[j+1]:
                    self.A[j],self.A[j+1] = self.A[j+1],self.A[j]

    def merge(self, other): # merging two lists
        result = MyList()
        for i in range(self.n):
            result.append(self.A[i])
        for i in range(len(other)):
            result.append(other[i])
        return result

    def slice(self, start, end):
        result = MyList()

        if start < 0:
            start = 0
        if end > self.n:
            end = self.n

        for i in range(start, end):
            result.append(self.A[i])

        return result


    # Encapsulated method __make_array     
    def __make_array(self,capacity):
        # referential array of c type
        return (capacity*ctypes.py_object)()
    

```

```python
L = MyList()
L.append(121)
L.append(122)
L.append(1)
L.append(33)
```
