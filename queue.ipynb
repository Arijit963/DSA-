```python
class Node :
    def __init__(self,value):
        self.data = value
        self.next = None
```

```python
# implementing queue with linked list

class Queue :
    def __init__(self):
        self.front = None
        self.rear = None

    def enqueue(self,val):

        new = Node(val)

        if self.rear == None: # in case queue is empty
            self.front = new
            self.rear = self.front
        else :
            self.rear.next = new # first we add the node next of rear
            self.rear = new  # then we assign the new node as rear

    def dequeue(self) : # deleteing from front

        if self.front == None :
            return "Empty queue"
        else :
            self.front = self.front.next 
    
    def is_empty(self):
        return self.front == None


    def traverse(self):
        temp = self.front
        while temp != None:
            print(temp.data)
            temp = temp.next

            
```

```python
q = Queue()
```

```python
q.enqueue(4)
q.enqueue(5)
q.enqueue(6)
```
